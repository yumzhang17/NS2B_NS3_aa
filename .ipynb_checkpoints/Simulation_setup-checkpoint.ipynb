1. Force field

CHARMM36m force field, stable for hairpin

**Note 1**

For CHARMM36m format PDB, we need to remove cap before transfer to gmx format

'''
dele atom sele type CAY .or. type HY1 .or. type HY2 .or. type HY3 .or. type CY .or. type OY -
     .or. type CAT .or. type HT1 .or. type HT2 .or. type HT3 .or. type NT .or. type HNT end
'''

Try box: 
"octahedron -d 1.0"

Final size: contain 32000 atoms. (1.0 may be better... didn't use 0.8 (28000 atoms)


Addions:

150 mM NaCl: 

N(w) : N (ions) = 1000g/18(g/mol) : 8.716g/(55.60g/mol) = n(w) : n(ions) = 370.37 : 1

We have 9800 waters in the system, means about 30 ions in the solvent.

Results:

30 Cl- / 37 Na+ to wt system

30 Cl- / 38 Na+ to 74A system 

30 Cl- / 36 Na+ to 81A system

30 Cl- / 39 Na+ to ns3m system

**Note 2**

When add restraints:

1) change topol* file:

add this to topol

'''

; Include Position restraint file

#ifdef POSRES

#include "posre_heavyatom_p74a.itp"

#endif

'''

add this to *.mdp file:

'''

define          = -DPOSRES      ; position restrain the protein

'''

2) Pay attention, two chains cannot add restraints together!!!!

i.e.,

load chain_a.itp

load posre_chain_a.itp

load chain_b.itp

load posre_chain_b.itp

....


So double check index file before make posre*.itp

and delete the last four lines in topol_*_chain_.itp file (which is initially to contain posres files for the whole system.


## 3. Temperature scanning

We had all systems @ 300 K first in directory:
/home/yumzhang/work/9-ns2b-ns3-aa/s6-prod
(pikes)

However, it seems like NS2B will not fall apart from NS3 even for the ns3m system within 100 ns.

(we checked restraints effect with running at 800 K in /home/yumzhang/work/9-ns2b-ns3-aa/test-800k)

Therefore, we did a temperature scanning for wt and ns3m systems, tempertaures are:

**350, 375, 400 425, 450, 475, 500, 600**

/home/yumzhang/work/9-ns2b-ns3-aa/s7-temp

(Final 600 is just for a top limitation scanning)

**Objective**

Find one optimal temperature conditions, where in wt system SN2B will stick to NS3 but escaping away in NS3m system.



### Conclusion 1.

Checked on 2022/12/28

Simulation time: around 60-80 ns. 

**wt**

600K: NS2B will dissociate within 50 ns. 
In all the other conditions, the binding between NS2B and NS3 is tight!

**NS3m**

Similarly as wt.

**Decisions**

Elongate the wt/ns3m 450, 475, 500 K three conditions to see the dissociation.

(in 475/500 cases, NS3m seems to dissociate a little bit around 300 ns)

### Conclusion 2.

Checked on 2023/01/02

Simulation time: 

wt: ~ 550 ns.

NS3m: 400 ns.

**wt**

No dissociation.

**NS3m**

450K: No dissociate within 400 ns.

**475K**: dissociate around 330 ns.

Unfold first, dissociate, try to rebind and refold, finially dissociate including 74 site.

**500K**: dissociate around 300 ns.

Unfold first, really rebind and refold (1163-1217 frame, 232-242 ns), keep stable for another 60 (1507, 301ns, re-dissociate, a unbind first then unfold). At around 360 ns, mid region dissociate.


**Decisions**

Elongate another 200 ns for ns3m 450K system? I don't think we need it. 400 ns is a long time and for the fully dissociate system we still see no dissociation. so... for the other two mutations, it's hard to see some interesting views within 500 ns (maybe?)

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from sklearn.decomposition import PCA
import scipy.stats as stats
import seaborn as sns
from scipy.stats import norm
from matplotlib.colors import LogNorm
from matplotlib import ticker, cm
from numpy import ma
from scipy.stats import gaussian_kde
import scipy.stats as st
import matplotlib.cm as cmap
import itertools  
````

### 4.1. 300 K 200 ns 

````python
%cd /home/yumzhang/pikes_work/9-ns2b-ns3-aa/s6-prod/
````

````python
a='wt'
b='p74a'
c='p81a'
d='ns3m'

for i in [a,b,c,d]:
    for j in range(1,3):
        top = md.load_pdb(str(i)+'/'+str(i)+'-prot.pdb')
        atom_sel=top.topology.select("chainid 1 and name CA")
        print(str(i),len(atom_sel))
        traj=md.load(str(i)+'/prod-trjconv.s'+str(j)+'.xtc',top=str(i)+'/'+str(i)+'-prot.pdb')
        ref=md.load(str(i)+'/'+str(i)+'-prot.pdb')
        rmsd=md.rmsd(traj,ref,0,atom_indices=atom_sel)
        plt.plot(np.arange(len(rmsd))/5,rmsd*10,label=str(i)+', s'+str(j))
        plt.legend(frameon=False,fontsize=12)

plt.ylim([0,5])
        
plt.ylabel('RMSD (A)',fontsize=16)
plt.xlabel('Time (ns)',fontsize=16)

plt.tick_params(axis='both', labelsize=16,direction='in',length=8,width=2)
plt.rcParams['axes.linewidth'] = 2
plt.rcParams["figure.figsize"] = (10,5)
        
    
plt.savefig('/home/yumzhang/pikes_work/9-ns2b-ns3-aa/results_23_01/300K_RMSD_4_sys_200ns.svg')
````

````python
a='wt'
b='p74a'
c='p81a'
d='ns3m'

for i in [a,b,c,d]:
    for j in range(1,3):
        top = md.load_pdb(str(i)+'/'+str(i)+'-prot.pdb')
        atom_sel=top.topology.select("chainid 1 and name CA")
        print(str(i),len(atom_sel))
        traj=md.load(str(i)+'/prod-trjconv.s'+str(j)+'.xtc',top=str(i)+'/'+str(i)+'-prot.pdb')

        dssp = md.compute_dssp(traj,simplified=True)
        sheet = np.where(dssp=='E', 1, 0)
        sheet = sheet[:,-38:]
        sheet = np.sum(sheet,0) / sheet.shape[0]

        plt.plot(51+np.arange(len(atom_sel)),sheet,label=str(i)+', s'+str(j))
        plt.legend(frameon=False,fontsize=12)

plt.ylim([0,1.1])
        
plt.ylabel('Probability',fontsize=16)
plt.xlabel('NS2B Residues',fontsize=16)

plt.tick_params(axis='both', labelsize=16,direction='in',length=8,width=2)
plt.rcParams['axes.linewidth'] = 2
plt.rcParams["figure.figsize"] = (10,5)
        
    
plt.savefig('/home/yumzhang/pikes_work/9-ns2b-ns3-aa/results_23_01/300K_2nd_4_sys_200ns.svg')
````
